```python
import aocd
from collections import defaultdict
import tqdm


test_input = """r, wr, b, g, bwu, rb, gb, br

brwrr
bggr
gbbr
rrbgbr
ubwu
bwurrg
brgr
bbrgwb"""
```

```python


def parse_input(puzzle_input):
    towels, patterns = puzzle_input.split("\n\n")
    towels = towels.replace(" ", "")
    towels = towels.split(",")
    patterns = patterns.split("\n")
    sorted_towels = list(reversed(sorted(towels, key=len)))
    towel_dict = defaultdict(list)
    for t in sorted_towels:
        towel_dict[t[0]].append(t)
    return patterns, towel_dict

def can_make(pattern, towel_dict, checked_patterns = None):
    if checked_patterns = None: #initialize some cache
        checked_patterns = {}

    if pattern in checked_patterns:
        return checked_patterns[pattern]
        
    if len(pattern) == 0:
        print("No colours left to match, solved!")
        checked_patterns[pattern] = True
        return True #we have parsed the whole thing...
    # print(f"Matching {pattern}")
    first_colour = pattern[0]
    towels_with_correct_prefix = towel_dict[first_colour]
    if len(towels_with_correct_prefix) == 0:
        #We cant find a towel starting with the correct colour, so we cant make this pattern.
        # print(f"No towels starting with {first_colour} in towel_dict")
        checked_patterns[pattern] = False
        return False
    else:
        #we have to try using all the towels.
        for towel in towels_with_correct_prefix:
            # print(f"solving {pattern} by trying {towel}"
            if pattern.startswith(towel):
                # print(f"Trying to solve {pattern} for {towel}")
                new_pattern = pattern[len(towel):]
                if can_make(new_pattern, towel_dict):
                    checked_patterns[pattern] = True
                    return True
                #     any_combo = True
    # print(f"Could not solve for {pattern}")
    return False



def solve1(puzzle_input):
    patterns, towel_dict = parse_input(puzzle_input)
    solvable_patterns = 0
    for p in patterns:
        print(f"\nSOLVING FOR PATTERN {p}")
        if can_make(p, towel_dict):
            solvable_patterns+=1
            # print(f"Pattern could be solved. Total patterns solved = {solvable_patterns}\n\n")
    return solvable_patterns
```

```python
def can_make_n_ways(pattern, towel_dict, checked_patterns = None):
    if checked_patterns == None: #initialize a cache
        checked_patterns = {}

    if pattern in checked_patterns:
        return checked_patterns[pattern]
        
    if len(pattern) == 0:
        # print("No colours left to match, solved!")
        checked_patterns[pattern] = 1 #one more way to solve the pattern?
        return 1 #we have parsed the whole thing...
    # print(f"Matching {pattern}")
    first_colour = pattern[0]
    towels_with_correct_prefix = towel_dict[first_colour]
    if len(towels_with_correct_prefix) == 0:
        #We cant find a towel starting with the correct colour, so we cant make this pattern.
        # print(f"No towels starting with {first_colour} in towel_dict")
        checked_patterns[pattern] = 0
        return 0
    else:
        #we have to try using all the towels.
        total_combos = 0
        for towel in towels_with_correct_prefix:
            # print(f"solving {pattern} by trying {towel}"
            if pattern.startswith(towel):
                # print(f"Trying to solve {pattern} for {towel}")
                new_pattern = pattern[len(towel):]
                #gather all the ways to make the pattern
                total_combos += can_make(new_pattern, towel_dict, checked_patterns)
                    # checked_patterns[pattern] = True
                    # return True
                #     any_combo = True
    # print(f"Could not solve for {pattern}")
    checked_patterns[pattern] = total_combos
    return total_combos
```

```python
patterns, towel_dict = parse_input(test_input)
total_combos = 0
for p in patterns:
    total_combos += can_make_n_ways(p, towel_dict)
print(total_combos)
```
