```python
import aocd
import numpy as np
```

```python
test_input = """#####
.####
.####
.####
.#.#.
.#...
.....

#####
##.##
.#.##
...##
...#.
...#.
.....

.....
#....
#....
#...#
#.#.#
#.###
#####

.....
.....
#.#..
###..
###.#
###.#
#####

.....
.....
.....
#....
#.#..
#.#.#
#####"""
```

```python
def parse_input(data):
    data = data.replace(".", "0")
    data = data.replace("#", "1")
    schematics = [np.array([[int(char) for char in line] for line in schematic.splitlines()]) for schematic in data.split("\n\n")]
    keys = []
    locks = []
    for s in schematics:
        if 1 in s[0]:
            locks.append(s)
        else:
            keys.append(s)
    return locks, keys
        
```

```python
locks, keys = parse_input(test_input)
```

```python
((locks[0] + keys[1]) == 2).any()
```

```python
def solve1(puzzle_input):
    locks, keys = parse_input(puzzle_input)
    fitting_pairs = set()
    for i, key in enumerate(keys):
        for j, lock in enumerate(locks):
            if not ((lock + key)==2).any():
                fitting_pairs.add((i,j))
    return len(fitting_pairs)
            
        
```

```python
solve1(test_input)
```

```python
real_input = aocd.get_data()
```

```python
solve1(real_input)
```
