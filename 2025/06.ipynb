```python
def parse_data(data):
    data = data.split("\n")
    # print(data)
    arr = [[int(x) for x in line.split(" ") if x!=""] for line in data[:-1]]
    arr = np.vstack(arr).T
    # print(arr)
    # print(data[-1])
    operators = data[-1].strip(" ").split()
    return arr, operators
```

```python
parse_data(test_data)
```

```python
def solve_part_1(data):
    grand_total = 0
    arrs, operators = parse_data(data)
    for arr, op in zip(arrs, operators):
        # print(arr, op)
        if op == "*":
            s = arr.prod()
        else:
            s = arr.sum()
        grand_total+=s
    return grand_total
            
        
        
```
