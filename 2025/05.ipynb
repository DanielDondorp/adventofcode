```python
def parse_data(data):
    ranges, ingredients = data.split("\n\n")
    ranges = ranges.split("\n")
    ranges = [[int(i) for i in r.split("-")] for r in ranges]
    ingredients = [int(i) for i in ingredients.split("\n")]
    return ranges, ingredients
```

```python
ranges, ingredients = parse_data(test_data)
```

```python
def find_fresh(data):
    ranges, ingredients = parse_data(data)
    fresh = []
    for ingredient in ingredients:
        for r in ranges:
            if ingredient >= r[0] and ingredient <= r[-1]:
                fresh.append(ingredient)
                break
    return fresh
```
