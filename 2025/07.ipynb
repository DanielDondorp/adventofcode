```python
def run_split_counter(arr, debug = "True"):
    split_counter = 0
    w, h = arr.shape
    for x, y in itertools.product(range(w), range(h)):
        if arr[x, y] == "^":
            # print(f"found ^ at {x, y}, above: {arr[x -1, y]}")
            if arr[x -1, y] == "|" or arr[x -1, y] == "S": #split beam:
                arr[x, y -1] = "|"
                arr[x, y+1] = "|"
                split_counter+=1
        elif arr[x, y] == "." and arr[x-1, y] == "|":
            arr[x,y] = "|"
        elif arr[x,y] == "S":
            arr[x+1, y] = "|"
        if y == 0 and debug:
            print_arr(arr)
    
    if debug:
        print_arr(arr)
        
    return split_counter
                    
        
```

```python
def solve_part_1(data, debug = True):
    arr = parse_data(data)
    return run_split_counter(arr, debug)
```

```python
def print_arr(arr):
    arr[arr == "0"] = " "
    a = "\n".join([" ".join(line) for line in arr.astype(str)])
    print(a+"\n")
```

```python
def parse_data(data):
    arr = np.array([[x for x in line] for line in data.split("\n") if "^" in line or "S" in line])
    # arr = np.pad(arr, pad_width=1)
    return arr.astype("object")
```

```python

def run_quantum_split_counter(arr, debug = "True"):
    # arr2 = np.zeros(arr.shape).astype(int)
    arr = arr.astype("object")
    arr[arr == "."] = "0"
    arr[arr == "S"] = "1"
    # arr2[arr == "S"] = 1
    # arr2 = np.pad(arr2, pad_width=1)
    h, w = arr.shape
    # print(w,h)
    for y, x in itertools.product(range(h), range(w)):
        if arr[y,x].isdigit() and arr[y-1, x] != "^": #propagate above-values for empty spaces
            arr[y,x] =  str(int(arr[y-1, x]) + int(arr[y,x]))
        if arr[y, x] == "^":
            if arr[y -1, x].isdigit(): #the splitter is hit by a beam, we propagate

                in_value = arr[y-1, x]
                #left:
                arr[y, x-1] = str( int(arr[y, x-1]) + int(in_value) )
                #right:
                arr[y, x+1] = str( int(arr[y, x+1]) + int(in_value) )
        if  x== 0 and debug:
            print_arr(arr.copy())
    
    if debug:
        print_arr(arr.copy())
        
    return arr
```

```python
def solve_part_2(data, debug = True):
    arr = parse_data(data)
    result = run_quantum_split_counter(arr.astype(str), debug)
    print(f"Total paths = {sum([int(x) for x in result[-1,:] if x.isdigit()])}")
```
