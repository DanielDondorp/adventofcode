```python
import aocd
import numpy as np
```

```python
test_data  = """987654321111111
811111111111119
234234234234278
818181911112111"""
```

```python
def parse_data(data):
    data = data.split("\n")
    data = [[int(x) for x in d] for d in data]
    return data
```

```python
test_batteries = parse_data(test_data)
```

```python
def find_largest_joltage(bank):
    start_index = np.argmax(bank[:-1])
    start_val = bank[start_index]
    end_val = max(bank[start_index+1:])
    jolts = int(str(start_val) + str(end_val))
    return jolts
    
    
    
```

```python
def solve_part_1(data):
    batteries = parse_data(data)
    total_jolts = 0
    for bank in batteries:
        total_jolts += find_largest_joltage(bank)
    print(f"{total_jolts =}")
```

```python
solve_part_1(test_data)
```

```python
def find_largest_joltage2(bank, n_batteries = 12, debug = True):
    batteries = ""
    for i in range(n_batteries, 0, -1):
        #define our available search space:
        #we have to have enough batteries downstream to construct our total of n_batteries
        if i <= 1:
            search_space = bank #we can use the entire s
        else:
            search_space = bank[:-(i-1)]
        
        
        #select the largest available battery in the correct seach space along with its index
        largest_battery = np.max(search_space)
        index_largest_battery = np.argmax(search_space)

        #append the largest battery value to our joltage
        batteries += str(largest_battery)

        #the new bank we can search through starts at the next index in our bank
        new_bank = bank[index_largest_battery+1:]
        
        
        if debug:
            print(f"{search_space=} derived from {bank=} with {i} batteries remaining, current batteries = {batteries}")
            # print(f"{bank=} \n{i} still needed so search space {bank} up to index {len(bank)} - {i} = {len(bank)-i}\n{search_space}")
            # print(f"{batteries=}, {len(batteries)=} new bank = {new_bank}\n")

        #set the new bank for the next cycle
        bank = new_bank
        #decrease our remaining required battery count:
        # n_batteries = n_batteries-1
    return int(batteries)

def solve_part_2(data):
    batteries = parse_data(data)
    total_jolts = 0
    for bank in batteries:
        total_jolts += find_largest_joltage2(bank, debug = False)
    print(f"{total_jolts =}")
```
