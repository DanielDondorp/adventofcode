```python
import aocd
import numpy as np
from functools import cache
```

```python
test_data = """aaa: you hhh
you: bbb ccc
bbb: ddd eee
ccc: ddd eee fff
ddd: ggg
eee: out
fff: out
ggg: out
hhh: ccc fff iii
iii: out"""
```

```python
def parse_data(data):
    graph = {}
    for line in data.split("\n"):
        k, vals = line.split(": ")
        graph[k] = vals.split(" ")
    return graph
```

```python
test_graph = parse_data(test_data)
```

```python
data = aocd.get_data()
```

```python
graph = parse_data(data)
```

```python
@cache
def count_paths(graph, start, goal, visited=set()):
    if start == goal:
        return 1
    visited.add(start)
    total = 0
    for nxt in graph.get(start, []):
        if nxt not in visited:
            total += count_paths(graph, nxt, goal, visited)
    visited.remove(start)
    return total
```

```python
def count_paths(graph, start, goal):
    @cache
    def cp(start, goal, visited = frozenset()):
        if start == goal:
            return 1
        visited.union(start)
        total = 0
        for nxt in graph.get(start, []):
            if nxt not in visited:
                total += cp(nxt, goal, visited)
        visited.difference(start)
        return total
    return cp(start, goal)
        
```

```python
count_paths(test_graph, "you", "out")
```

```python
count_paths(graph, "you", "out")
```
