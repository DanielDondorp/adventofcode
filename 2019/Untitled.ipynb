```python
import numpy as np
import pandas as pd
```

```python
puzzle_input = "246540-787419"
```

```python
low = 246540
high = 787419
```

# part 1

```python
all_vals = np.arange(low, high)
#make dataframe
df = pd.DataFrame(all_vals, columns=["value"])
#create column that records the diffs in a number
df["differential"] = df.value.apply(lambda x: np.diff([int(i) for i in str(x)]))

#any number that never decreases and has at least one instance of 0 (no diff means adjacent numbers are the same)
df["valid"] = df.differential.apply(lambda x: ((x >= 0).all() and (x == 0).any()))

#give us the number of valid numbers
df["valid"].sum()
```

# part2

```python
valid_numbers = df[df["valid"]]
```

```python
d2 = pd.Series([x for x in np.diff(np.vstack(valid_numbers["differential"]), axis = 1)])
```

```python
d3 = pd.Series([x for x in np.diff(d2) ])
```

```python
valid_numbers["d2"] = valid_numbers["differential"].apply(lambda x: np.diff(x))
```

```python
valid_numbers["d3"] = valid_numbers["d2"].apply(lambda x: np.diff(x))
```

```python
valid_numbers.head()
```

```python
v = 0
for i, row in valid_numbers.iterrows():
    if (row["d2"] == 0).any() and (row["d3"] != 0).all():
        print(row["value"],row["differential"], row["d2"], row["d3"])
        v+=1
```

```python
valid_passwords = []
for value in valid_numbers["value"]:
    pw = str(value)
    groups = [pw.count(c) for c in set(pw)]
    if any([x==2 for x in groups]):
        valid_passwords.append(value)
```
