```python
import numpy as np
import aocd
```

```python
puzzle_input = aocd.get_data(day = 3, year = 2019)
```

```python
test_wire1 = "R8,U5,L5,D3"
test_wre2 = "U7,R6,D4,L4"
```

```python
directions = {"R": [0, 1],
              "U": [-1, 0],
              "D": [1, 0],
              "L": [0,-1]}
              
```

```python
def wire_to_coords(wire):
    steps = wire.split(",")
    wire_run = []
    for step in steps:
        d = step[0]
        n = int(step[1:])
        wire_run += int(n) * [directions[d]]
    return np.cumsum(np.vstack([[0,0], wire_run]), axis = 0)
                
                
        
```

```python
r1 = wire_to_coords(test_wire1)
r2 = wire_to_coords(test_wre2)
```

```python
overlay = r1[(r1 == r2).all(axis = 1)]
```

```python
def manhattan_distance(loc):
    return np.sum(abs(loc))
```

```python
manhattan_distance(overlay[1:])
```

```python
def solve1(data):
    wire1, wire2 = data.split("\n")
    r1 = wire_to_coords(wire1)
    r2 = wire_to_coords(wire2)
    r1s = set([tuple(x) for x in r1])
    r2s = set([tuple(x) for x in r2])
    crossings = r1s.intersection(r2s)
    crossings = [np.array(x) for x in crossings]
    dists = [manhattan_distance(loc) for loc in crossings if sum(loc)!=0]
    return min(dists)
```

```python
solve1(test_wire1+"\n"+test_wre2)
```

```python
def solve2(data):
    w1, w2 = data.split("\n")    
    c1 = wire_to_coords(w1)
    c2 = wire_to_coords(w2)
    r1s = set([tuple(x) for x in c1])
    r2s = set([tuple(x) for x in c2])
    crossings = r1s.intersection(r2s)
    crossings = [tuple(x) for x in crossings if not (x[0] == 0 and x[1] == 0)]
    t1 = [tuple(x) for x in c1]
    t2 = [tuple(x) for x in c2]
    dists = []
    for crossing in crossings:
        steps1 = t1.index(crossing)
        steps2 = t2.index(crossing)
        dists.append(steps1+steps2)
    return np.min(dists)
```

```python
test1 = "R75,D30,R83,U83,L12,D49,R71,U7,L72\nU62,R66,U55,R34,D71,R55,D58,R83"
```

```python
solve2(test1)
```

```python
solve2(puzzle_input)
```
